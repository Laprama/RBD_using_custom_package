#### Generate Synthetic Data

```python
# Function to create samples with Gaussian noise
def create_samples(base_gradient, num_samples):
    height, width = base_gradient.shape
    # Create an empty array to hold all samples
    samples = np.zeros((num_samples, height, width))
    
    # Generate samples by adding Gaussian noise to the base gradient
    for i in range(num_samples):
        noise = np.random.normal(0, 0.2, (height, width))  # Mean 0, Stddev 0.1
        samples[i] = base_gradient + noise
    return samples

# Base gradients
height, width = 64, 157
class1_base = np.tile(np.linspace(0, 1, width), (height, 1))
class2_base = np.tile(np.linspace(1, 0, width), (height, 1))

# Generate samples for Class 1 and Class 2
class1_samples = create_samples(class1_base, 10)
class2_samples = create_samples(class2_base, 10)
```

```python
class ConvolutionalNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1,6,4,1)
        self.conv2 = nn.Conv2d(6, 16, 4, 1)
        
        self.fc1 = nn.Linear(16*3*8, 200)
        self.fc2 = nn.Linear(200, 60)
        self.fc3 = nn.Linear(60,1)
        
    def forward(self, X):
        X = F.relu(self.conv1(X))
        X = F.max_pool2d(X,4,4)
        X = F.relu(self.conv2(X))
        X = F.max_pool2d(X,4,4)
        
        X = X.view(-1, 16*3*8)
        
        X = F.relu(self.fc1(X) )
        X = F.relu(self.fc2(X))
        X = F.relu(self.fc3(X))
        
        # return F.log_softmax(X, dim=1 )
        return X
```

```python
import time

start_time = time.time()

epochs = 10
train_losses = []
test_losses = []
train_correct = []
test_correct = []

for i in range(epochs):
    trn_corr = 0
    tst_corr = 0
    
    # Run the training batches
    for b, (X_train, y_train) in enumerate(train_loader):
        b+=1
        
        # Apply the model
        y_pred = model(X_train)  # we don't flatten X-train here
        loss = criterion(y_pred, y_train.unsqueeze(1).float())
 
        # Tally the number of correct predictions
        predicted = torch.max(y_pred.data, 1)[1]
        batch_corr = (predicted == y_train).sum()
        trn_corr += batch_corr
        
        # Update parameters
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    train_losses.append(loss)
    train_correct.append(trn_corr)
        
        
print(f'\nDuration: {time.time() - start_time:.0f} seconds') # print the time elapsed            
```
