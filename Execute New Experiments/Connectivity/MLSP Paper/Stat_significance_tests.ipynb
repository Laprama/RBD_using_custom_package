```python
# Function to check if any two columns are the same
def check_identical_columns(df):
    columns = df.columns
    n = len(columns)
    identical_columns = []

    for i in range(n):
        for j in range(i + 1, n):
            if df[columns[i]].equals(df[columns[j]]):
                identical_columns.append((columns[i], columns[j]))

    return identical_columns

check_identical_columns(eeg_type_dist)
```
