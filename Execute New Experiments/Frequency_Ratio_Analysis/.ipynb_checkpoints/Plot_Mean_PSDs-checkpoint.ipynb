#### Manipulate df for plotting

```python
#Plot one region G
plt.figure(figsize=(14,20),dpi=200)

num_to_classes = {0: 'HC', 1: 'PD', 2: 'PD+RBD', 3: 'RBD'}
df_region = df[region_features + ['label'] ]

for val in np.unique(df_region['label']):
    if val in [0,1,2]:
        mean_psd = df_region[ df_region['label'] == val].drop(columns = ['label']).mean(axis = 0)
        plt.plot(np.arange(0.5,40.125, 0.125) , mean_psd.values , label = num_to_classes[val])
    
plt.legend()
```
